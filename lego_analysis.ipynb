```python
# imports and set up dataframes
import pandas as pd

df = pd.read_csv('https://raw.githubusercontent.com/BrandonSpadafora/lego-analysis/master/datasets/lego_sets.csv')
parent_theme = pd.read_csv('https://raw.githubusercontent.com/BrandonSpadafora/lego-analysis/master/datasets/parent_themes.csv')
```

## Task #1: What percentage of all licensed sets ever released were Star Wars Themed?

```python
#merging both the dfs
merged = df.merge(parent_theme, left_on='parent_theme', right_on='name')
merged.head()
```

```python
#dropping un-needed column
merged.drop(columns='name_y', inplace=True)
merged.head()
```

```python
#getting all the licensed themes and dropping NAs
licensed = merged[merged['is_licensed']]
licensed = licensed.dropna(subset=['set_num'])
licensed.head()
```

```python
#getting just star wars themes
star_wars = licensed[licensed['parent_theme']== 'Star Wars']
star_wars.head(50)
```

```python
#getting the percent of Star Wars sets out of all the sets
star_war_pot = int(star_wars.shape[0]/licensed.shape[0]*100) #pot = percent of total it's an abbreviation
print(star_war_pot)
```

51% of all licensed sets ever released were star wars

# Task #2: In which year was Star Wars not the most popular licensed theme?

```python
#sorting licensed sets by year
licensed_sorted = licensed.sort_values('year')
licensed_sorted['count'] = 1
licensed_sorted.head()
```

```python
#grouping licensed_sorted by year and parent_theme
summed_df = licensed_sorted.groupby(['year','parent_theme']).sum().reset_index()
summed_df.head()
```

```python
#Getting the count of what set had the most releases each year
max_df = summed_df.sort_values('count', ascending=False).drop_duplicates(['year'])
max_df.sort_values('year', inplace=True)
max_df
```

In 2017 Super Heroes was more popular than Star Wars

# Task #3 how many sets were released each year (1950-2017)?

```python
#getting rid of null values in the set_num column
cleaned_df = merged.dropna(subset=['set_num'])
cleaned_df.head()
```

```python
#getting the number of sets released each year
cleaned_df['count'] = 1
sets_per_year = cleaned_df.groupby(['year']).sum().reset_index()[['year','count']]
sets_per_year.head(50)
```

```python
#it's the same as the previous cell, except that it's easier to see every year
for index,row in sets_per_year.iterrows():
    print(row['year'], row['count'])
```
